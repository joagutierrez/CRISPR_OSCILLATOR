```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.integrate import odeint
```

Differential equation: Driven, damped harmonic oscillator
$$ \frac{\partial^2 a}{\partial t^2} = -k a -c \frac{\partial a}{\partial t} + F(t)   $$
$$ v(t) = \frac{\partial a}{\partial t} $$
$$ \dot a(t) = v(t) $$
$$ \dot v(t) = -k a(t) -c v(t) + F(t) $$
$$ \mathbf{y}(t) = (a(t), v(t))$$

```python
# Differential Equation

def harmOSC(y, t, k, c):
    a, v = y
    dadt = v
    dvdt = -k*a - c*v + F(t)
    return [dadt,dvdt]
```

```python
# Parameters

T = 200   # sec
increments = 1001
t = np.linspace(0, T, increments)

k = 5
c = 0.1
def F(t):
    omega0 = 0.5
    A = 0.01
    return A*np.cos(omega0*t)
```

```python
# Initial Conditions

a0 = 0.05
v0 = 0.1
y0 = [a0,v0]
```

```python
# Run

y = odeint(harmOSC, y0, t, args=(k, c))
```

```python
plt.plot(t,y[:,0])
```
